```python
SEED = 1
```

```python
import pandas as pd
```

```python
train_df = pd.read_csv('Dataset-train-vf.csv',index_col=0)#,on_bad_lines='skip')
test_df = pd.read_csv('Dataset-test-vf.csv',index_col=0)#,on_bad_lines='skip')
```

```python
def dataInfoPrinter():
    display(train_df.info())
    display(test_df.info())
    display(train_df.describe())
    display(test_df.describe())
```

```python
display(train_df)
display(test_df)
```

```python
dataInfoPrinter()
```

**As we can see, both column x2 and x6 has NaNs; however, x2 has a lot of missing values so that it cannot be filled with averages nor removing some rows. Because of that, we will drop column x2.**

**For column x6, we have 2 ways, either average or drop rows. It is prefered to add the average as the data is not large and having more training is better**

```python
# drop x2
train_df = train_df.drop(['x2'], axis=1)
test_df = test_df.drop(['x2'], axis=1)
# add average to missing rows in x6
# Mean imputation
train_df['x6'] = train_df['x6'].fillna(train_df['x6'].mean())
test_df['x6'] = test_df['x6'].fillna(test_df['x6'].mean())
# Median imputation
#df['x6'].fillna(df['x6'].median(), inplace=True)
```

```python
dataInfoPrinter()
```

**As we can see above, the column x14 is catagorical, so we need to convert it to numirical. we will be using one-hot-encoding. And Also convert y to**

```python
train_df = pd.get_dummies(train_df,columns=['x14'], prefix_sep='.', dummy_na=False)#pd.get_dummies(train_df,columns=train_df.iloc[:,:-1].columns, prefix_sep='.', dummy_na=False)#.fillna(0)
test_df = pd.get_dummies(test_df,columns=['x14'], prefix_sep='.', dummy_na=False) #pd.get_dummies(test_df ,columns=test_df.iloc[:,:-1].columns, prefix_sep='.', dummy_na=False)#.fillna(0) ,columns=test_df['x14'] pd.concat([test_df.drop('x14', axis=1), pd.get_dummies(test_df['x14'], prefix='x14', prefix_sep='.')], axis=1)
train_df['y'] = train_df['y'].map({'Low': 0, 'High': 1})
test_df['y'] = test_df['y'].map({'Low': 0, 'High': 1})
```

```python
# Get a list of all column names except 'y'
cols = [col for col in train_df.columns if col != 'y']
# Add 'y' at the end of the list
cols.append('y')
# Reindex the DataFrame with the new column order
train_df = train_df[cols]
# Get a list of all column names except 'y'
cols = [col for col in test_df.columns if col != 'y']
# Add 'y' at the end of the list
cols.append('y')
# Reindex the DataFrame with the new column order
test_df = test_df[cols]
```

```python
dataInfoPrinter()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming you have handled missing values appropriately
scaler = MinMaxScaler()

# Select numerical columns (exclude categorical and target columns)
numerical_cols = train_df.columns[train_df.dtypes != 'object']
numerical_cols = numerical_cols.drop('y')  # Exclude the target if it's encoded

# Fit and transform training data
train_df[numerical_cols] = scaler.fit_transform(train_df[numerical_cols])

# Transform test data using the same scaler
test_df[numerical_cols] = scaler.transform(test_df[numerical_cols])
```

```python
dataInfoPrinter()
```

**As we can see bellow, we have class imbalance which causes many problems to a lot of classifiers. Becuase of that, we will apply SMOTE oversampling only on the training data.**

```python
display(train_df["y"].value_counts(normalize=True))
display(test_df["y"].value_counts(normalize=True))
```

```python
from imblearn.over_sampling import SMOTE
X_train = train_df.drop('y', axis=1)
y_train = train_df['y']

smote = SMOTE(random_state=SEED)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
train_df = pd.concat([X_train_smote, y_train_smote], axis=1)
display(train_df["y"].value_counts(normalize=True))
```

# **Part B:**

```python
selected_train_samples = train_df.sample(n=50, random_state=SEED)
train_df_active_learning = train_df.copy(deep=True)
```

```python
def getNonNaN(df):
    # DataFrame where y is not NaN
    df_not_nan = df[df['y'].notna()]
    return df_not_nan
def getNanNaN(df):
    df_nan = df[df['y'].isna()]
    return df_nan
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
old_part_A_acc = 0.8938271604938272
print(f"Part A Log Reg Accuracy is = {old_part_A_acc}\n")
X_test = test_df.drop('y', axis=1)
y_test = test_df['y']
# Assume X_train, y_train, X_test, y_test are predefined
# Initialize with 50 labeled data points
# n_initial = 50
# initial_idx = np.random.choice(range(len(X_train)), size=n_initial, replace=False)
# X_labeled, y_labeled = df_not_nan.iloc[:,:-1], df_not_nan.iloc[:,-1]
# X_unlabeled, y_unlabeled = df_nan.iloc[:,:-1], df_nan.iloc[:,-1]

# X_labeled, y_labeled = X_train[initial_idx], y_train[initial_idx
# X_unlabeled, y_unlabeled = np.delete(X_train, initial_idx, axis=0), np.delete(y_train, initial_idx, axis=0)

# Initialize the Logistic Regression model
model = LogisticRegression(C= 0.5, n_jobs= 8, penalty= None, solver= 'sag',random_state=SEED)
train_df_least_confidence = train_df_active_learning.copy(deep=True)
n_queries = 1188 - 50  # This is quite high, consider reducing if performance is a concern
for i in range(n_queries):
    df_not_nan = train_df_least_confidence[train_df_least_confidence['y'].notna()]
    df_nan = train_df_least_confidence[train_df_least_confidence['y'].isna()]
    X_labeled, y_labeled = df_not_nan.iloc[:, :-1], df_not_nan.iloc[:, -1]
    X_unlabeled, y_unlabeled = df_nan.iloc[:, :-1], df_nan.iloc[:, -1]
    # print(X_labeled.shape)
    # print(y_labeled.shape)
    # print(X_unlabeled.shape)
    # print(y_unlabeled.shape)

    # Fit model
    model.fit(X_labeled, y_labeled)

    # Predict probabilities on the unlabeled pool
    probas = model.predict_proba(X_unlabeled)

    # Select the sample with the highest uncertainty
    query_idx = np.argmin(np.abs(probas.max(axis=1) - 0.5))

    # Retrieve the actual index from df_nan to use in train_df_active_learning
    actual_query_idx = df_nan.iloc[query_idx].name  # Using `.name` to get the original index from df_nan

    # Assume the oracle or mechanism to label the new sample
    # Here it should correctly refer to train_df_active_learning to update the label
    # This line should be corrected to something that mimics the labeling process
     # Update the label in 'train_df_active_learning' using the actual index
    if actual_query_idx in train_df.index:
        train_df_least_confidence.at[actual_query_idx, 'y'] = train_df.at[actual_query_idx, 'y']
    accuracy = accuracy_score(y_test, model.predict(X_test))
    #print(f'Accuracy after query {i+1}: {accuracy:.4f}')
    # Progress output for monitoring
    if i % 100 == 0:  # Print progress every 100 iterations
        print(f"Iteration {i}: Sample {actual_query_idx} labeled.")
        print(f'Accuracy after query {i+1}: {accuracy:.4f}')
    # Evaluate and print accuracy

    if accuracy >= old_part_A_acc:
        break
print(f'Accuracy after query {i+1}: {accuracy:.4f}')
```
